```python
import pandas as pd
from keybert import KeyBERT
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Load the CSV data
data = pd.read_csv('untitled.csv')

# Create a new column for the extracted keywords
data['Keywords'] = ''

# Initialize the KeyBERT model
kw_model = KeyBERT()

# Process each review text and extract keywords
for index, row in data.iterrows():
    review_text = row['Review Text']
    keywords = kw_model.extract_keywords(review_text, keyphrase_ngram_range=(1, 2), stop_words='english',
                                         use_maxsum=True, nr_candidates=20, top_n=10)
    data.at[index, 'Keywords'] = ', '.join([kw for kw, _ in keywords])
```

```python
# Save the updated data to a new CSV file
data.to_csv('untitled (extracted).csv', index=False)
```
